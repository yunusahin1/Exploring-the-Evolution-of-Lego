```python
# Question to answer
'''1. What percentage of all licensed sets ever released were Star Wars themed?
Save your answer as a variable the_force in the form of an integer (e.g. 25).

2. In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?
Save your answer as a variable new_era in the form of an integer (e.g. 2012).
'''
```

```python
# Import pandas and read in the DataFrame, and inspect it
import pandas as pd
lego_sets = pd.read_csv('datasets/lego_sets.csv')
lego_sets.info()
```

```python
# Drop relevant missing rows
lego_sets_clean = lego_sets.dropna(subset=['set_num', 'name', 'theme_name'])
lego_sets_clean.info()
```

```python
# Get list of licensed sets
parent_themes = pd.read_csv('datasets/parent_themes.csv')
licensed_themes = parent_themes[parent_themes['is_licensed'] == True]['name']
licensed_themes.head()
```

```python
# Subset for licensed sets
licensed = lego_sets_clean['parent_theme'].isin(licensed_themes)
licensed_sets = lego_sets_clean[licensed]
licensed_sets.head()
```

```python
# Calculate the percentage of licensed sets that are Star Wars themed
all_sets = len(licensed_sets)
star_wars_sets = len(licensed_sets[licensed_sets['parent_theme'] == 'Star Wars'])
ratio = star_wars_sets/all_sets
the_force = int(ratio*100)
print(the_force)
```

```python
# Create a pivot table of sets released by theme per year
licensed_pivot = licensed_sets.pivot_table(index = 'year', columns = 'parent_theme', values = 'set_num', aggfunc = 'count')
licensed_pivot.head()
```

```python
# Find the year when Star Wars was not the top theme
new_era = licensed_pivot[licensed_pivot['Star Wars'] < licensed_pivot.max(axis='columns')].index[0]
print(new_era)
```
